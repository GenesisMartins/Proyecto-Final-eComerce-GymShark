```python
# Definir una función para extraer los valores numéricos de la columna Price.
def extract_price(value):
    # Buscar todos los números después del símbolo del euro
    numbers = re.findall(r'\d+', value)
    # Convertir los números a enteros y devolver el primero encontrado
    if numbers:
        return int(numbers[0])
    else:
        return None

# Aplicar la función a la columna "Price"
gymshark_products['Price'] = gymshark_products['Price'].apply(extract_price)
```

```python
# Función para generar códigos de stock aleatorios
def generate_stock_code():
    return ''.join([str(random.randint(0, 9)) for _ in range(5)]) + random.choice('ABCDEFGHIJKLMNOPQRSTUVWXYZ')

# Agregar la columna "StockCode" al DataFrame gymshark_products
gymshark_products['StockCode'] = gymshark_products.apply(lambda _: generate_stock_code(), axis=1)
```
